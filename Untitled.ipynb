```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import streamlit as st
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import japanize_matplotlib

# ãƒ•ã‚¡ã‚¤ãƒ«å
DATA_FILE = "saitama_data.csv"
RANKING_FILE = "team_ranking.csv"

# ãƒ‡ãƒ¼ã‚¿èª­ã¿è¾¼ã¿
@st.cache_data
def load_data():
    return pd.read_csv(DATA_FILE)

df = load_data()
numeric_columns = df.select_dtypes(include=['float64', 'int64', 'int32']).columns.tolist()

# UI: ã‚¿ã‚¤ãƒˆãƒ«ã¨é¸æŠ
st.title("ğŸ“Š åŸ¼ç‰çœŒã‚ªãƒ¼ãƒ—ãƒ³ãƒ‡ãƒ¼ã‚¿åˆ†æã‚¢ãƒ—ãƒª")
st.markdown("2ã¤ã®æŒ‡æ¨™ã‚’é¸ã³ã€å¯è¦–åŒ–ã¨å˜å›å¸°åˆ†æã‚’ä½“é¨“ã§ãã¾ã™ã€‚")

# ãƒ‡ãƒ¼ã‚¿é¸æŠ
x_col = st.selectbox("Xè»¸ã«ã™ã‚‹é …ç›®", numeric_columns)
y_col = st.selectbox("Yè»¸ã«ã™ã‚‹é …ç›®", [col for col in numeric_columns if col != x_col])

# ã‚°ãƒ©ãƒ•ã®ç¨®é¡
graph_type = st.selectbox("è¡¨ç¤ºã™ã‚‹ã‚°ãƒ©ãƒ•ã®ç¨®é¡", [
    "æ•£å¸ƒå›³", "æŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•", "æ£’ã‚°ãƒ©ãƒ•", "å††ã‚°ãƒ©ãƒ•", "ãƒ’ã‚¹ãƒˆã‚°ãƒ©ãƒ ", "ç®±ã²ã’å›³"
])

st.subheader("ğŸ“ˆ ã‚°ãƒ©ãƒ•ã®è¡¨ç¤º")

# ã‚°ãƒ©ãƒ•è¡¨ç¤º
fig = plt.figure()
if graph_type == "æ•£å¸ƒå›³":
    sns.scatterplot(data=df, x=x_col, y=y_col)
elif graph_type == "æŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•":
    plt.plot(df[x_col], df[y_col], marker="o")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
elif graph_type == "æ£’ã‚°ãƒ©ãƒ•":
    plt.bar(df[x_col], df[y_col])
elif graph_type == "å††ã‚°ãƒ©ãƒ•":
    if "å¸‚ç”ºæ‘" in df.columns:
        df_sorted = df.sort_values(y_col, ascending=False).head(10)
        plt.pie(df_sorted[y_col], labels=df_sorted["å¸‚ç”ºæ‘"], autopct='%1.1f%%')
    else:
        st.warning("å¸‚ç”ºæ‘åˆ—ãŒãªã„ãŸã‚ã€å††ã‚°ãƒ©ãƒ•ãŒæç”»ã§ãã¾ã›ã‚“ã€‚")
elif graph_type == "ãƒ’ã‚¹ãƒˆã‚°ãƒ©ãƒ ":
    sns.histplot(df[x_col], kde=True)
elif graph_type == "ç®±ã²ã’å›³":
    sns.boxplot(data=df[[x_col, y_col]])

st.pyplot(fig)

# å˜å›å¸°åˆ†æ
st.subheader("ğŸ“Š å˜å›å¸°åˆ†æ")

X = df[[x_col]].dropna()
y = df[y_col].dropna()
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

st.write(f"å›å¸°å¼ï¼š y = {model.coef_[0]:.3f} * x + {model.intercept_:.3f}")
st.write(f"æ±ºå®šä¿‚æ•° RÂ²ï¼š {r2:.3f}")

# å›å¸°ç·šä»˜ãæ•£å¸ƒå›³ï¼ˆPlotlyï¼‰
fig2 = px.scatter(df, x=x_col, y=y_col, trendline="ols", title="å›å¸°ç›´ç·šä»˜ãæ•£å¸ƒå›³")
st.plotly_chart(fig2)

# ãƒ©ãƒ³ã‚­ãƒ³ã‚°ç™»éŒ²
st.subheader("ğŸ† ãƒãƒ¼ãƒ ãƒ©ãƒ³ã‚­ãƒ³ã‚°æ©Ÿèƒ½")

team_name = st.text_input("ãƒãƒ¼ãƒ åã‚’å…¥åŠ›ã—ã¦ãã ã•ã„")

if st.button("ãƒ©ãƒ³ã‚­ãƒ³ã‚°ã«ç™»éŒ²"):
    if team_name:
        new_record = pd.DataFrame([{
            "ãƒãƒ¼ãƒ å": team_name,
            "X": x_col,
            "Y": y_col,
            "R2": r2
        }])

        if os.path.exists(RANKING_FILE):
            existing = pd.read_csv(RANKING_FILE)
            updated = pd.concat([existing, new_record], ignore_index=True)
        else:
            updated = new_record

        updated.to_csv(RANKING_FILE, index=False)
        st.success("ãƒ©ãƒ³ã‚­ãƒ³ã‚°ã«ç™»éŒ²ã—ã¾ã—ãŸï¼")

# ãƒ©ãƒ³ã‚­ãƒ³ã‚°è¡¨ç¤º
if os.path.exists(RANKING_FILE):
    st.subheader("ğŸ“‹ ãƒãƒ¼ãƒ ãƒ©ãƒ³ã‚­ãƒ³ã‚°ä¸€è¦§ï¼ˆRÂ²é †ï¼‰")
    df_rank = pd.read_csv(RANKING_FILE).sort_values("R2", ascending=False)
    st.dataframe(df_rank)
```
