```python
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import random

stack_usage = []
heap_usage = []
MAX_TIME = 50

fig, ax = plt.subplots()

def simulate_usage(i):
    if random.random() < 0.6:
        stack = stack_usage[-1] + random.randint(1, 3) if stack_usage else 5
    else:
        stack = max(0, stack_usage[-1] - random.randint(1, 3)) if stack_usage else 5

    if random.random() < 0.5:
        heap = heap_usage[-1] + random.randint(2, 4) if heap_usage else 10
    else:
        heap = max(0, heap_usage[-1] - random.randint(2, 4)) if heap_usage else 10

    stack_usage.append(stack)
    heap_usage.append(heap)

    if len(stack_usage) > MAX_TIME:
        stack_usage.pop(0)
        heap_usage.pop(0)

    ax.clear()
    ax.plot(stack_usage, label='スタック領域', color='pink')
    ax.plot(heap_usage, label='ヒープ領域', color='skyblue')
    ax.set_ylim(0, 50)
    ax.set_title("スタック領域とヒープ領域の使用量（相対値）")
    ax.set_xlabel("時刻")
    ax.set_ylabel("使用量")
    ax.legend(loc='upper left')

ani = FuncAnimation(fig, simulate_usage, interval=300)
HTML(ani.to_jshtml())  # ここで表示
```
